```python
df['Comment']=df['Comment'].str.lower()
df['Label']=df['Label'].str.lower()

#Cleaning
#menghilangkan web
df['Comment']=df['Comment'].str.replace('(https?://[\w\.\/]*)', '')
#menghilangkan encoding simbol
df['Comment']=df['Comment'].str.replace('(?:&(?:lt|nbsp|amp|gt);)', '')
#menghilangkan tag @ dan hastag
df['Comment']=df['Comment'].str.replace('(@|#)\w+', '')
#menghilangkan karakter lain selain A-Z, a-z, 0-9
df['Comment']=df['Comment'].str.replace('[^A-Za-z0-9\s\-\/]', '')
#menghilangkan tanda khusus
df['Comment']=df['Comment'].str.replace('(\-|\/)', ' ')
#menghilangkan enter, break, new line
df['Comment']=df['Comment'].str.replace('\n', ' ')
#menghilangkan karakter khusus
df['Comment']=df['Comment'].str.replace("\.\.", " ")

#menghilangkan data kosong
df['Label'].replace('', np.nan, inplace=True)
df['Comment'].replace('', np.nan, inplace=True)

#menghapus baris yang kosong
df.dropna(subset=['Comment'], inplace=True)
df.dropna(subset=['Label'], inplace=True)

#menghapus baris duplikat
df = df.drop_duplicates()
```

```python
text = df["stemmer"].tolist()
print(text)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('indonesian')
stop.extend(('.', '', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','/','-','#','&','***'))

frequency_words = {}
for data in text:
    data_list = data.split(' ')
    print(data_list)
    for token in data_list:
        token = token.rstrip()
        token = token.lstrip()
        if token not in stop:
            if token in frequency_words:
                count = frequency_words[token]
                count = count + 1
                frequency_words[token] = count
            else:
                frequency_words[token] = 1
```

```python
import math
import operator
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud()
wordcloud.generate_from_frequencies(frequency_words)
plt.figure(figsize=(14,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
#tokenizing : mengindex kata menjadi int
from tensorflow.keras.preprocessing.text import Tokenizer

token = Tokenizer()
token.fit_on_texts(text)
hasiltoken = token.index_word
print(hasiltoken)
```

```python
encode_token = token.texts_to_sequences(text)
print(encode_token)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_kata = 40
X = pad_sequences(encode_token, maxlen = max_kata, padding="post")
X
```

```python
y = df['Label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=40, test_size = 0.3, stratify=y)

X_train = np.asarray(X_train)
X_test = np.asarray(X_test)
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, ComplementNB, BernoulliNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
model = ComplementNB()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
model = BernoulliNB()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV
np.random.seed(555)

cv_method = RepeatedStratifiedKFold(n_splits=10,  n_repeats=3, random_state=555)

classifier1 = GaussianNB()
params_nb1 = {'var_smoothing': np.logspace(0,-9, num=100)}
gs_NB = GridSearchCV(estimator=classifier1, param_grid=params_nb1,verbose=1,cv=5, scoring='accuracy')
```

```python
gs_NB.fit(X_train, y_train)
gs_NB.score(X_test, y_test)
```

```python
y_pred_gsn = gs_NB.predict(X_test)

print('Train Accuracy : %.3f'%gs_NB.best_estimator_.score(X_train, y_train))
print('Test Accuracy : %.3f'%gs_NB.best_estimator_.score( X_test, y_test))
print('Best Accuracy Through Grid Search : %.3f'% gs_NB.best_score_)
print('Best Parameters : ',gs_NB.best_params_)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
cm1 = confusion_matrix(y_test, y_pred_gsn)
cm1
```
